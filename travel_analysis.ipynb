```python
# change budget level to ordinal variate
df['budget_level'] = df['budget_level'].map({'Budget': 1, 'Mid-range': 2, 'Luxury': 3})


# converting region to a 0/1 binary variate
region_dummies = pd.get_dummies(df['region'], prefix='region', dtype=int)
# add to df
df = pd.concat([df, region_dummies], axis=1)

print(df.columns)
df.head()
```

```python
# change the json field to a python list
import json
df['ideal_durations'] = df['ideal_durations'].apply(json.loads)
```

```python
# create a new column for indicating if the places are ideal for a weekend trip
df['is_weekend_trip'] = df['ideal_durations'].apply(lambda x: 1 if isinstance(x, list) and 'Weekend' in x else 0)
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
import math
```

```python
# extract the fields that are used in the model
df_model = df[['region_africa', 'region_asia',
       'region_europe', 'region_middle_east', 'region_north_america',
       'region_oceania', 'region_south_america', 'budget_level',
       'culture', 'adventure', 'nature', 'beaches', 'nightlife', 'cuisine',
       'wellness', 'urban', 'seclusion', 'is_weekend_trip']]

print(df_model.columns)
df_model.head()
```

```python
# set the predictor variables and the predicted value
X = df_model[['region_africa', 'region_asia',
       'region_europe', 'region_middle_east', 'region_north_america',
       'region_oceania', 'region_south_america', 'budget_level',
       'culture', 'adventure', 'nature', 'beaches', 'nightlife', 'cuisine',
       'wellness', 'urban', 'seclusion']]
y = df_model['is_weekend_trip']
```

```python
# split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=23)
print(len(X_train))
print(len(y_train))
print(len(X_test))
print(len(y_test))
```

```python
# feature scaling
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
classifier = KNeighborsClassifier(n_neighbors=int(math.sqrt(len(df))), p=2,metric='euclidean')
classifier.fit(X_train, y_train)
```

```python
# predict the results of the test set
y_pred = classifier.predict(X_test)
print(y_pred)
print(len(y_pred))
```

```python
# evaluate model
cm = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:\n",cm)
print("F1-Score: ", f1_score(y_test, y_pred))
print("Accuracy Score: ", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
accuracy_scores = []
k_range = range(1, 30)
for k in k_range:
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(X_train, y_train)
    acc_score = accuracy_score(y_test, model.predict(X_test))
    accuracy_scores.append(acc_score)

plt.plot(k_range, accuracy_scores, marker='o')
plt.title('Accuracy vs k')
plt.xlabel('k')
plt.ylabel('Accuracy')
plt.xticks(k_range)
plt.grid(True)
plt.show()
```

Let's try with 14 neighbours now since the graph seems to be stabilising a bit starting from that point.

```python
classifier = KNeighborsClassifier(n_neighbors=14, p=2,metric='euclidean')
classifier.fit(X_train, y_train)

# predict the results of the test set
y_pred = classifier.predict(X_test)
y_pred

# evaluate model
cm = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:\n",cm)
print("F1-Score: ", f1_score(y_test, y_pred))
print("Accuracy Score: ", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

The reported accuracy score and F1-score are higher in this case than with 23 neighbours.

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# set the predictor variables and the predicted value
X = df_model[['region_africa', 'region_asia',
       'region_europe', 'region_middle_east', 'region_north_america',
       'region_oceania', 'region_south_america', 'budget_level',
       'culture', 'adventure', 'nature', 'beaches', 'nightlife', 'cuisine',
       'wellness', 'urban', 'seclusion']]
y = df_model['is_weekend_trip']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=23)
print(len(X_train))
print(len(y_train))
print(len(X_test))
print(len(y_test))
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=23)
rf_classifier.fit(X_train, y_train)

y_pred = rf_classifier.predict(X_test)
```

```python
# evaluate model
cm = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:\n",cm)
print("F1-Score: ", f1_score(y_test, y_pred))
print("Accuracy Score: ", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

We see that both the F1-score and the accuracy score improve with the Random Forest model than the K-Nearest Neighbours model.
